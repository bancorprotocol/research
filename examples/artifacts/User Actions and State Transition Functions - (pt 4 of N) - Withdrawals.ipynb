#### Review from last session

Configure the system state.


```python
# Initialize the main simulator class
v3 = BancorDapp(cooldown_time=7)

# whitelist tokens
v3.whitelist_token('usdc')
v3.whitelist_token('usdt')
v3.whitelist_token('dai')

# Create the users
v3.create_user('Alice')

# Set the initial starting balances
v3.set_user_balance(user_name='Alice', tkn_name='eth',tkn_amt=1000000)
v3.set_user_balance(user_name='Alice', tkn_name='link',tkn_amt=1000000)
v3.set_user_balance(user_name='Alice', tkn_name='bnt',tkn_amt=1000000)
v3.set_user_balance(user_name='Alice', tkn_name='usdc',tkn_amt=1000000)
v3.set_user_balance(user_name='Alice', tkn_name='usdt',tkn_amt=1000000)
v3.set_user_balance(user_name='Alice', tkn_name='dai',tkn_amt=1000000)

v3.create_user('Bob')

# Set the initial starting balances
v3.set_user_balance(user_name='Bob', tkn_name='eth',tkn_amt=1000000)
v3.set_user_balance(user_name='Bob', tkn_name='link',tkn_amt=1000000)
v3.set_user_balance(user_name='Bob', tkn_name='bnt',tkn_amt=1000000)
v3.set_user_balance(user_name='Bob', tkn_name='usdc',tkn_amt=1000000)
v3.set_user_balance(user_name='Bob', tkn_name='usdt',tkn_amt=1000000)
v3.set_user_balance(user_name='Bob', tkn_name='dai',tkn_amt=1000000)

v3.describe()
```

#### Review from last session

Create some deposits to bootstrap the pools.

```python
# Create deposits
timestamp = 0
v3.deposit('usdc', '500000', 'Alice', timestamp=timestamp)
v3.deposit('eth', '500000', 'Alice', timestamp=timestamp)
v3.deposit('link', '500000', 'Bob', timestamp=timestamp)

# Iterate timestamp
timestamp = 1
v3.deposit('link', '500000', 'Bob', timestamp=timestamp)

# Enable trading
v3.enable_trading('link', timestamp=timestamp)
v3.enable_trading('eth', timestamp=timestamp)
v3.enable_trading('usdc', timestamp=timestamp)
v3.describe()
```
